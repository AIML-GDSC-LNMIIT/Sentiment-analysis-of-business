```python
class StarXReview :
    def __init__(self, stars, review) :
        self.stars = stars
        self.review = review
    def __str__(self):
        print(self.stars, self.review)
        return ""

dictionary = {}
for idx, row in df.iterrows() :
    myList = dictionary.get(row['business_id'], [])
    myList.append(StarXReview(row['stars'], row['review']))
    dictionary[row['business_id']] = myList
```

```python
df_new = pd.DataFrame({"business" : [], "avg_stars" : [], "review" : []})
df_new
```

```python
for key, value in dictionary.items() :

    avg_star_count = 0
    for i in value :
        avg_star_count = avg_star_count + i.stars
    avg_star_count = avg_star_count / len(value)

    review = ""
    for i in value :
        review = review + " "+ i.review

    df_new.loc[len(df_new)] = {"business" : key, "avg_stars" : avg_star_count, "review" : review}
```

```python
df_new['business'] = df_new['business'].apply(lambda x : x.replace("/", ""))
```

```python
for idx, row in df_new.iterrows() :
    text = row['review']
    lower_case = text.lower()
    cleaned_text = lower_case.translate(str.maketrans('', '', string.punctuation))
    score = SentimentIntensityAnalyzer().polarity_scores(cleaned_text)
    df_new.at[idx, 'Sentiment_Score_neg'] = score['neg']
    df_new.at[idx, 'Sentiment_Score_pos'] = score['pos']
    if score['neg'] > score['pos']:
        df_new.at[idx, 'Sentiment_Score_remark'] = "Negative Sentiment"
    elif score['neg'] < score['pos']:
        df_new.at[idx, 'Sentiment_Score_remark'] = "Positive Sentiment"
    else:
        df_new.at[idx, 'Sentiment_Score_remark'] = "Neutral Sentiment"
```
